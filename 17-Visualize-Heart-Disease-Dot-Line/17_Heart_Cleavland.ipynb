```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Reading the dataset
df = pd.read_csv("Cleavland.csv", header=None)
df.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
```

```python
print(df)
```

```python
# Data Cleaning
df = df.replace('?', np.nan)
df = df.dropna()
```

```python
print(df)
```

```python
df.isnull().sum().sum()
```

```python
df.shape
```

```python
coli=["chol","fbs","restecg"]
df[coli].hist(bins=10)
plt.show()
```

```python
# Creating two subsets
subset1 = df[df['sex'] == 0]  # Subset with sex 0 (presumably female)
subset2 = df[df['sex'] == 1]  # Subset with sex 1 (presumably male)

# Plotting the histograms for both subsets
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(subset1['age'], bins=10, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution - Female')

plt.subplot(1, 2, 2)
plt.hist(subset2['age'], bins=10, color='lightgreen')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution - Male')
plt.show()
```

```python
plt.plot(df["age"],df["cp"],"ro")
plt.show()
```

```python
# Plotting the dot plots
plt.figure(figsize=(8, 6))
sns.stripplot(x='cp', y='trestbps', data=df, jitter=True)
plt.xlabel('Chest Pain Type')
plt.ylabel('Resting Blood Pressure')
plt.title('Chest Pain vs. Resting Blood Pressure')
plt.show()
```

```python
#try
plt.figure(figsize=(30,30))
df["age"].value_counts().plot(kind="bar")
plt.legend()
plt.show()
```

```python
# Plotting the bar plots
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
subset1['num'].value_counts().plot(kind='bar', color='lightblue')
plt.xlabel('Heart Disease')
plt.ylabel('Frequency')
plt.title('Heart Disease Distribution - Female')

plt.subplot(1, 2, 2)
subset2['num'].value_counts().plot(kind='bar', color='lightgreen')
plt.xlabel('Heart Disease')
plt.ylabel('Frequency')
plt.title('Heart Disease Distribution - Male')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['sex'].value_counts().plot(kind='bar', color='pink')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.title('Sex Distribution')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.lineplot(data=df,x="age",y="cp")
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.lineplot(data=df,x="thalach",y="slope")
plt.show()
```

```python
# Plotting the line charts
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(subset1['thalach'], color='purple')
plt.xlabel('Index')
plt.ylabel('Heart Rate')
plt.title('Heart Rate Over Time - Female')

plt.subplot(1, 2, 2)
plt.plot(subset2['thalach'], color='orange')
plt.xlabel('Index')
plt.ylabel('Heart Rate')
plt.title('Heart Rate Over Time - Male')
plt.show()
```

```python
asd=df["cp"].value_counts()
plt.figure(figsize=(10,5))
plt.pie(asd,labels=asd.index,autopct="%1.1f%%")
plt.legend()
plt.show()
```

```python
# Plotting Boxplot with Histogram and Scatterplot
plt.figure(figsize=(14, 6))

# Boxplot
plt.subplot(1, 3, 1)
sns.boxplot(y='age', data=df, orient='v')
plt.title('Boxplot - Age')

# Histogram
plt.subplot(1, 3, 2)
sns.histplot(df['age'], kde=True, color='skyblue')
plt.title('Histogram - Age')

# Scatterplot
plt.subplot(1, 3, 3)
sns.scatterplot(x='age', y='chol', data=df, hue='num')
plt.title('Scatterplot - Age vs. Cholesterol')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Adding boxplots for age on the side
sns.boxplot(x='age', data=df, orient='v', width=0.2, color='lightgray')

plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Scatterplot with Boxplots')
plt.legend(title='Heart Disease', loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.regplot(x='age', y='chol', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.title('Scatterplot with Regression Line - Age vs. Cholesterol')
plt.show()
```

```python
plt.figure(figsize=(14, 6))
sns.regplot(x='thalach', y='slope', data=df, scatter_kws={'color': 'green'}, line_kws={'color': 'purple'})
plt.xlabel('Maximum Heart Rate Achieved')
plt.ylabel('Slope of ST Segment')
plt.title('Line Chart with Regression Line - Maximum Heart Rate vs. Slope of ST Segment')
plt.show()
```
