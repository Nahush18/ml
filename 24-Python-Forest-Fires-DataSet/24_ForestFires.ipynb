```python
from sklearn.linear_model import LinearRegression


X = df[['X', 'Y','FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain']]
y = df['area']

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)

# Plotting regression line for one feature at a time (assuming 'temp')
plt.scatter(df['temp'], y, color='blue')
plt.plot(df['temp'], model.predict(X), color='red')
plt.xlabel('Temperature')
plt.ylabel('Area')
plt.title('Linear Regression')
plt.show()
```

```python
# Select only numeric columns
numeric_cols = df.select_dtypes(include=np.number)

# Calculate mean and standard deviation for each numeric feature
means = numeric_cols.mean(axis=0)
std_devs = numeric_cols.std(axis=0)

# Calculate Z-scores manually
z_scores = (numeric_cols - means) / std_devs

# Define threshold for outliers
threshold = 3

# Get indices of outliers
outlier_indices = np.where(np.abs(z_scores) > threshold)

# Remove outliers from original DataFrame
clean_df = df.drop(outlier_indices[0])

# Define X and y again
X_clean = clean_df[['X', 'Y', 'FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain']]
y_clean = clean_df['area']

# Fit linear regression model to cleaned data
model = LinearRegression()
model.fit(X_clean, y_clean)

# Plotting regression line for one feature at a time (assuming 'temp')
plt.scatter(clean_df['temp'], y_clean, color='blue')
plt.plot(clean_df['temp'], model.predict(X_clean), color='red')
plt.xlabel('Temperature')
plt.ylabel('Area')
plt.title('Linear Regression (Without Outliers)')
plt.show()
```

```python
from sklearn.metrics import r2_score

# R-squared with outliers
r2_with_outliers = model.score(X, y)

# R-squared without outliers
r2_without_outliers = model.score(X_clean, y_clean)

print("R-squared with outliers:", r2_with_outliers)
print("R-squared without outliers:", r2_without_outliers)
```
