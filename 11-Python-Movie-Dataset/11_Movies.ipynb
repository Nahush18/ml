# Casting

```python
Casting = pd.pivot_table(DataFrame, index=['id', 'original_language'], values=['original_title', 'spoken_languages'], aggfunc='count')
Casting
```

```python
plt.figure(figsize=(8, 6))
plt.hist(Casting['original_title'], bins=20, color='green', alpha=0.7)
plt.title('Histogram of Original Titles in Casting DataFrame')
plt.xlabel('Original Title Count')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram for each parameter
DataFrame.hist(figsize=(12, 10))
plt.suptitle('Histograms for Parameters')
plt.show()
```

```python
# Bar graph for original_language counts
plt.figure(figsize=(20, 4))
sns.countplot(data=DataFrame, x='original_language')
plt.title('Number of Movies per Original Language')
plt.xlabel('Original Language')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Heatmap for correlation matrix
# Exclude non-numeric columns
numeric_df = DataFrame.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
